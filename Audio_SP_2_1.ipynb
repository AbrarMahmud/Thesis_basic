#Audio applications
(Using Transformers Libraries)

**Audio classification:** easily categorize audio clips into different categories .that is whether a recording is of a barking dog or a meowing cat, or what music genre a song belongs to.

**Automatic speech recognition:** transform audio clips into text by transcribing them automatically.

**Speaker diarization:** Use Trnasformers to identify speakers

**Text to speech:** create a narrated version of a text that can be used to produce an audio book, help with accessibility.



#Audio classification:

```python
# Syntax for installing libraries in Google-Colab
!pip install datasets[audio]
```

```python
from datasets import load_dataset
from datasets import Audio

minds = load_dataset("PolyAI/minds14",name="en-AU",split="train")
minds = minds.cast_column("audio",Audio(sampling_rate = 16_000))
```

An example of classification using **pipeline**

```python
from transformers import pipeline

classifier = pipeline(
    "audio-classification",
    model="anton-l/xtreme_s_xlsr_300m_minds14",
)

```

```python
tmp_data = minds[300]

## let us observe an output
print(classifier(tmp_data["audio"]["array"]))

## verify it using the bulit in data (intent_class)-that helps us
## to know the true intent of the speech
id2label = minds.features["intent_class"].int2str
id2label(tmp_data["intent_class"])
```

```python
from transformers import pipeline
pipe = pipeline("text-to-speech",model="suno/bark-small")
```

```python
text = "Ladybugs have had important roles in culture and religion, being associated with luck, love, fertility and prophecy. "
output = pipe(text)
```
